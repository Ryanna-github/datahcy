# 数据清洗-评论内容

- 结果保存在 `../data/reply/reply_by_bvid_clean2`，该文件夹已清洗完评论相关内容，但其中包含评论者个人信息（在用户信息数据清洗和整合过程需要用）
- 运行过 `clenser_user.ipynb` 后会得到 `../data/reply/reply_by_bvid_clean_simple` 即不包含用户个人信息的数据，个人信息已提取到 `../data/user/` 对应文件中

```python
import pandas as pd
import numpy as np
import re
import os
import time
import datetime
```

```python
path_r = "../data/reply/reply_by_bvid_raw"
path_r2 = "../data/reply/reply_by_bvid_clean"
replies = os.listdir(path_r)
vid_list = [t[6:18] for t in replies]
```

```python
def comment_member(m):
    if isinstance(m, str):
        m = eval(m)
    info_keys = ['mid', 'uname', 'sex', 'sign', 'level']
    info_values = [m[key] for key in ['mid', 'uname', 'sex', "sign"]]
    info_values.append(m['level_info']['current_level'])
    return dict(zip(info_keys, info_values))

def comment_content(m):
    if isinstance(m, str):
        m = eval(m)
        return {"message": m['message']}
```

```python
fails = []
for file in replies:
    try:
        df = pd.read_csv(path_r + "\\" + file)
        # 删除冗余
        df = df.drop(['mid', 'type', 'root', 'dialog', 'count', 'rcount', 'state', 'fansgrade', 'attr', 'rpid_str'], axis = 1)
        df = df.drop(['parent', 'root_str', 'parent_str', 'action', 'folder', 'up_action', 'show_follow', 'invisible', 'reply_control', 'replies', 'assist'], axis = 1)
        
        # 评论者个人信息
        info_add_member = [comment_member(t) for t in df['member']]
        df_add_member = pd.DataFrame(info_add_member)

        # 评论内容
        info_add_content = [comment_content(t) for t in df['content']]
        df_add_content = pd.DataFrame(info_add_content)

        # 时间
        df['date'] = [datetime.datetime.strptime(time.ctime(t), '%a %b %d %H:%M:%S %Y').strftime('%m/%d/%Y') for t in df['ctime']]
        df['time'] = [datetime.datetime.strptime(time.ctime(t), '%a %b %d %H:%M:%S %Y').strftime('%H:%M:%S') for t in df['ctime']]

        df = pd.concat([df, df_add_member, df_add_content], axis = 1).drop(['member', 'content'], axis = 1)
        new_path = path_r2 + "\\" + "_clean_".join(file.split('_'))
        print(new_path)
        df.to_csv(new_path, index=False, header=True, encoding='UTF-8')
    except:
        print("fail at", file)
        fails.append(file)
```
